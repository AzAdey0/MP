```python
from Bio.PDB import PDBParser, PDBIO
```

```python
parser=PDBParser(QUIET=True)
io=PDBIO()
# Load structures
protein = parser.get_structure("protein", "HDE-unrelaxed.pdb")
heme = parser.get_structure("hemeA", "hemeA.pdb")

# Example: print chains to confirm
for model in heme:
    for chain in model:
        print("Chain:", chain.id)
```

```python
heme_residue = list(heme.get_residues())[0]
```

```python
heme_residue = None
for model in heme:
    for chain in model:
        for residue in chain:
            if residue.get_resname().strip() in ["HEM", "HEME", "HEC"]:
                heme_residue = residue
                break
```

```python
print("Heme residue:", heme_residue)
```

```python
monomer_chain=list(protein[0])[0]
heme_residue.id = ("H_HETATM", 201, " ")  # Assign a unique ID to avoid clashes
monomer_chain.add(heme_residue)
```

```python
print("Residues in monomer after insertion:")
for res in monomer_chain:
    print(res.get_resname(), res.id)
```

```python

io.set_structure(protein)
io.save("HDE-unrelaxed_wheme.pdb")
```

```python
with open("af2_aligned.pdb") as f1, open("hemeA.pdb") as f2, open("af2_with_heme.pdb", "w") as out:
    monomer_lines = [line for line in f1 if line.startswith(("ATOM", "HETATM", "TER"))]
    heme_lines = [line for line in f2 if line.startswith(("HETATM", "TER"))]
    out.writelines(monomer_lines + heme_lines + ["END\n"])
```
